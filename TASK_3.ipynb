```python
# Import Libraries
import pandas as pd
import sqlite3
from google.colab import files
# Upload Dataset
uploaded = files.upload()
```

```python
# Load Cleaned Dataset
df = pd.read_excel("cleaned_data.xlsx")
print("Dataset Loaded Successfully")
print(df.head())
```

```python
# Create SQLite Database
conn = sqlite3.connect("sales_data.db")

# Store Data in SQL Table
df.to_sql("orders", conn, if_exists="replace", index=False)

print("\nDatabase Created Successfully")
```

```python
# 1. SELECT Query
# -----------------------------------

query1 = """
SELECT *
FROM orders
LIMIT 5
"""

result1 = pd.read_sql(query1, conn)

print("\nFirst 5 Records")
print(result1)
```

```python
# 2. WHERE Query
# -----------------------------------

query2 = """
SELECT OrderID,
       Product,
       TotalPrice
FROM orders
WHERE OrderStatus = 'Delivered'
"""

result2 = pd.read_sql(query2, conn)

print("\nDelivered Orders")
print(result2.head())
```

```python
# 3. ORDER BY Query
# -----------------------------------

query3 = """
SELECT Product,
       TotalPrice
FROM orders
ORDER BY TotalPrice DESC
"""

result3 = pd.read_sql(query3, conn)

print("\nHighest Value Orders")
print(result3.head())
```

```python
# 4. COUNT Query
# -----------------------------------

query4 = """
SELECT COUNT(*) AS TotalOrders
FROM orders
"""

result4 = pd.read_sql(query4, conn)

print("\nTotal Orders")
print(result4)
```

```python
# 5. AVG Query
# -----------------------------------

query5 = """
SELECT AVG(TotalPrice) AS AveragePrice
FROM orders
"""

result5 = pd.read_sql(query5, conn)

print("\nAverage Order Value")
print(result5)
```

```python
# 6. SUM Query
# -----------------------------------

query6 = """
SELECT SUM(TotalPrice) AS TotalSales
FROM orders
"""

result6 = pd.read_sql(query6, conn)

print("\nTotal Sales")
print(result6)
```

```python
# 7. GROUP BY Query
# -----------------------------------

query7 = """
SELECT Product,
       COUNT(*) AS ProductCount
FROM orders
GROUP BY Product
"""

result7 = pd.read_sql(query7, conn)

print("\nProduct Counts")
print(result7)
```

```python
# 8. HAVING Query
# -----------------------------------

query8 = """
SELECT Product,
       COUNT(*) AS ProductCount
FROM orders
GROUP BY Product
HAVING COUNT(*) > 1
"""

result8 = pd.read_sql(query8, conn)

print("\nProducts Appearing More Than Once")
print(result8)
```

```python
# 9. GROUP BY + SUM
# Business Insight
# -----------------------------------

query9 = """
SELECT Product,
       SUM(TotalPrice) AS TotalRevenue
FROM orders
GROUP BY Product
ORDER BY TotalRevenue DESC
"""

result9 = pd.read_sql(query9, conn)

print("\nRevenue by Product")
print(result9)
```

```python
# 10. GROUP BY + AVG
# -----------------------------------

query10 = """
SELECT Product,
       AVG(TotalPrice) AS AverageRevenue
FROM orders
GROUP BY Product
ORDER BY AverageRevenue DESC
"""

result10 = pd.read_sql(query10, conn)

print("\nAverage Revenue by Product")
print(result10)
```

```python
# Close Connection
# -----------------------------------

conn.close()

print("\nSQL Data Analysis Completed Successfully!")
```
